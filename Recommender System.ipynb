### UBCF (User Based Collaborative Filtering)
- User-Based Collaborative Filtering is a technique used to predict the items that a user might like on the basis of ratings given to that item by the other users who have similar taste with that of the target user.

### Metrics 1: Correlation Matrix

```python
ubcf_data = pd.pivot_table(data=movie,values='rating',index='movie',columns='userId').fillna(0)
ubcf_data.columns = movie['userId'].unique()
ubcf_data
```

```python
ubcf_data.corr().round(2)
```

## Metric 2: Euclideon Distance

```python
from sklearn.metrics import pairwise_distances
```

```python
ubcf_euc = pd.pivot_table(data=movie,values='rating',index='userId',columns='movie').fillna(0)
ubcf_euc.index = movie['userId'].unique()
ubcf_euc
```

```python
euc_dis = pd.DataFrame(data=pairwise_distances(X=ubcf_euc,Y=None,metric='euclidean')).round(2)
euc_dis.index = movie['userId'].unique()
euc_dis.columns = movie['userId'].unique()
euc_dis
```

```python
movie[(movie['userId'] == 8) |(movie['userId'] == 16) | (movie['userId'] == 6979) | (movie['userId'] == 7087)]
```

### Metric 3 : Cosine Distance

```python
pairwise_distances(X=ubcf_euc,Y=None, metric='cosine')
```

```python
cos_euc = pd.DataFrame(data=1-pairwise_distances(X=ubcf_euc,Y=None,metric='cosine')).round(2)
cos_euc.index = movie['userId'].unique()
cos_euc.columns = movie['userId'].unique()
cos_euc
```
